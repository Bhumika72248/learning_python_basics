## set 
## creating a set

```python
a={"animals", "balls","cat"}
print(a)
print(type(a))
```

## set items are unordered

```python
a={"animals", "balls","cat"}
print(a[0])
```

## duplicates not allowed

```python
a={"animals", "balls","cat", "balls", "cat", "cat"}
print(a)
```

## True and 1 is considered the same value


```python
a={"animals", "balls","cat", "animals", 1,2,3,True}
print(a)
```

```python
a={"animals", "balls","cat", "animals", 1,2,3,True}
print(a)
print(len(a))
```

## Join two sets

```python
x={"a","b","c"}
y={1,2,3,4,5}
z=x.union(y)
print(z)
```

```python
## write a python program to find the max nad min values
setn={5,10,15,2,50}
print("print original set")
print(setn)
print("maximum value os the setn is")
print(max(setn))
print("minimum value os the setn is")
print(min(setn))
```

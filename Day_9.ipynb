```python
a={"name" : "priya",
   "age" : "26",
   "country" : "india"}
a['age']=28
print(a)
```

```python
a={"name" : "priya",
   "age" : "26",
   "country" : "india", "gender" : "female"}
a.update({"mail id" : "priya@gmail.com"})
print(a)
```

```python
a={"name" : "priya",
   "age" : "26",
   "country" : "india"}
a.pop("age")
print(a)
```

```python
a={"name" : "priya",
   "age" : "26",
   "country" : "india"}
del a["age"]
print(a)
```

```python
a={"name" : "priya",
   "age" : "26",
   "country" : "india"}
del a
print(a)
```

```python
a={"name" : "priya",
   "age" : "26",
   "country" : "india"}
a.clear()
print(a)
print(type(a))
```

## check whether a given key already exist in a dictionary 

```python
a={
    "name" : "priya",
   "age" : "26",
   "country" : "india"
 }
def is_key_present(x):
    if x in a:
        print("key is present in the dictionary")
    else:
        print("key is not present in the dictionary")
is_key_present("name")
is_key_present("shopping")
is_key_present("age")
```

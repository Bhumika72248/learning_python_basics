## slicing in strings

```python
s="Hello students how are you ?"
print(s[0:5])
```

```python
s="Hello students how are you ?"
print(s[6:18])
```

```python
s="Hello students how are you ?"
print(s[0:20:2]) #if i want to skip one number then i have to give 2 but if i want to skip two number then i have to  give 3
```

```python
# reversing the string
s="Hello students how are you ?"
print(s[::-1])
```

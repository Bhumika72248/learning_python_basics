## Bar graph

```python
x=[1,2,3,4]
y=[5,6,7,8]
plt.bar(x,y)
plt.show()
```

```python
## changing colours 
x=[1,2,3,4]
y=[5,6,7,8]
c=['r','b','g','y']
plt.bar(x,y,color=c)
plt.show()
```

```python
## for horizontal representatioin
import numpy as np
x=np.array(['A','B','C','D'])
y=np.array([3,4,5,10])
plt.barh(x,y)
plt.show()
```

```python
## wanting all the bars are same colour 
import numpy as np
x=np.array(['A','B','C','D'])
y=np.array([3,4,5,10])
plt.bar(x,y, color="violet")
plt.show()
```

```python
## we can change colour through color code
import numpy as np
x=np.array(['A','B','C','D'])
y=np.array([3,4,5,10])
plt.barh(x,y, color="#4CAF50")
plt.show()
```

```python
## we can change its width in vertical bars
import numpy as np
x=np.array(['A','B','C','D'])
y=np.array([3,4,5,10])
plt.bar(x,y,width=0.2)
plt.show()
```

```python
## we can change its height in horizontal bars
import numpy as np
x=np.array(['A','B','C','D'])
y=np.array([3,4,5,10])
plt.barh(x,y,height=0.1)
plt.show()
```

## scatter plot

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
y=[3,29,2,9,4,6]
plt.scatter(x,y)
plt.show()
```

```python
## i want to give title
x=[1,2,3,4,5,6]
y=[3,29,2,9,4,6]
plt.scatter(x,y)
plt.xlabel("year")
plt.ylabel("month")
plt.title("scatter")
plt.show()
```

```python
## i want to give title
x = [1,2,3,4,5,6]
y = [3,29,2,9,4,6]
c = ["red","black","yellow","orange","purple","grey"]
plt.scatter(x,y,color=c)
plt.xlabel("year")
plt.ylabel("month")
plt.title("scatter")
plt.show()
```

```python
## i want adjust its size, s=....
x = [1,2,3,4,5,6]
y = [3,29,2,9,4,6]
c = ["red","black","yellow","orange","purple","grey"]
plt.scatter(x,y,color=c, s=150)
plt.xlabel("year")
plt.ylabel("month")
plt.title("scatter")
plt.show()
```

## matplotlib pie chart

```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40]
y=['english','hindi', 'maths', 'science']
plt.pie(x,labels=y)
plt.show()
```

```python
x=[10,20,30,40]
y=['english','hindi', 'maths', 'science']
c=['yellow','pink','black','magenta']
plt.pie(x,labels=y, colors=c)
plt.legend() ##for labels
plt.show() 
```

## Concept of loops

## While loop
program to display number form 1 to 5

```python
i=1
n=5
while i<=n:
    print(i)
    i=i+1
```

enter the values from user and make a multiplication table till 10

```python
num=int(input("enter a number for which you want to print table from"))
i=1
n=10
while i<=n:
    print(i,"*",num," = ",num*i)
    i=i+1
print("this is the multiplication")
```

## For loop

```python
fruits=["apple","banana","cherry"]
for x in fruits:
    print(x)
```

```python
def electricity_bill(n):
    if n<=500:
        print("your bill is Rs",n*5)
    elif n>500 and n<=700:
        print("your bill is Rs", n*10)
    elif n>700 and n<=1000:
        print("your bill is Rs", n*15)
    elif n>1000:
        print("your bill is Rs", n*20)
```

## Tuple in python

```python
## creating a tuple
my_tuple=(1,2,3)
print(my_tuple)
print(type(my_tuple))
```

## Mixed Data types

```python
my_tuple=(1,"hello",3,2,4+5j)
print(my_tuple)
print(type(my_tuple))
```

## empty tuple

```python
my_tuple=()
print(my_tuple)
print(type(my_tuple))
```

## nested tuple

```python
a=("data",[1,2,3],(4,5,6))
print(a)
print(type(a))
```

## Creating tuple with only one item

```python
var1=("hello")
print(var1)
print(type(var1))
```

```python
## when we want to create single item tuple then use comma(,) after single value like string or int etc
var2=("hello",)
print(var2)
print(type(var2))
```

## tuple constructor

```python
a=tuple(("a","b","c"))
print(a)
print(type(a))
```

## indexing

```python
a=(47,48,45,46,48,4,8)
print(a[5])
```

```python
a=(47,48,45,46,48,4,8,"bhumi","aashi")
print(a[7])
print(a[-3])
```

## slicing

```python
a=(47,48,45,46,48,4,8)
print(a[0:6])
```

```python
a=(47,48,45,46,48,4,8)
print(a[:])
```

```python
a=(47,48,45,46,48,4,8)
print(a[::2])
```

```python
a=(47,48,45,46,48,4,8)
print(a[:6])
```

```python
a=(47,48,45,46,48,4,8)
print(a[::-1])
```

```python
t=(1,2,3,4,5,6,7,8)
print(t[::])
print(t[3:])
print(t[:4])
print(t[-2:5])
```

```python
a=(25,26,27,28,29)
y=list(a)
y.append(30)
a=tuple(y)
print(a)
```

```python
a=(25,26,27,28,29)
y=list(a)
y.remove(26)
a=tuple(y)
print(a)
```

## loop

```python
this_tuple=("a","b","c")
for x in this_tuple:
    print(x)
```

## join two tuple

```python
tuple1=("a","b","c")
tuple2=(1,2,3)
tuple3=tuple1+tuple2
print(tuple3)
```

## multiply

```python
a=("apple","bhumi","aashi")
b=a*2
print(b)
```

```python
t=(1,2,3,4,5,6,7,8)
print(t*2)
print(t+t+t)
print(t+t)
print(len(t))
```
